```python
# By Ismael Valenzuela (@aboutsecurity / @thinkredactblue)

import numpy as np
import matplotlib.pyplot as plt

# Example IOC data: each entry has:
# [source_reliability, indicator_prevalence, freshness_factor, relevance_weight]
ioc_data = {
    "IOC A": [0.9, 10, 0.8, 1.0],
    "IOC B": [0.7, 50, 0.9, 1.2],
    "IOC C": [0.8, 5, 0.4, 0.8],
    "IOC D": [0.6, 20, 0.7, 1.1]
}

scores = {}
for ioc, factors in ioc_data.items():
    source_reliability, indicator_prevalence, freshness_factor, relevance_weight = factors
    
    # Example formula:
    # IOC_Score = (Source Reliability) * (1 / Indicator Prevalence)
    #             * (Freshness Factor) * (Relevance Weight)
    score = (source_reliability 
             * (1 / indicator_prevalence) 
             * freshness_factor 
             * relevance_weight)
    scores[ioc] = score

# Prepare data for visualization
iocs = list(scores.keys())
ioc_scores = list(scores.values())

# Create a bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(iocs, ioc_scores, color='skyblue')

# Annotate the bar chart with the actual score values
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, 
             f"{height:.4f}", 
             ha='center', va='bottom')

plt.title("IOC Scoring Demonstration")
plt.xlabel("IOC")
plt.ylabel("Score (arbitrary units)")
plt.ylim(0, max(ioc_scores) * 1.2)  # add a bit of space at the top

plt.show()
```
