```python
# By Ismael Valenzuela (@aboutsecurity / @thinkredactblue)
# Based on SANS SEC568 - Combating Supply Chain Attacks with Product Security Testing
# https://www.sans.org/cyber-security-courses/combating-supply-chain-attacks-product-security-testing/ 

# https://towardsdatascience.com/graph-visualisation-basics-with-python-part-iii-directed-graphs-with-graphviz-50116fb0d670
# https://h1ros.github.io/posts/introduction-to-graphviz-in-jupyter-notebook/ 

import graphviz
f = graphviz.Digraph(filename = "output/attacktree.gv")
names = ["1","2","3","4","5"]
positions = ["Attacker leverages CVE-202x-xxxx" + "\\n" +"to remotely exploit kernel",
             "Attacker gains root access to UDM Pro", "Attacker does MiTM on traffic", "Attacker captures valid creds", 
             "Attacker has valid creds providing long term access"]
for name, position in zip (names, positions):
     f.node(name, position, shape = "record", style = "filled", color = "skyblue")
 
#Specify edges
f.edge("1","2");
f.edge("2","3");
f.edge("3","4");
f.edge("4","5");

f
```
