```python
# By Ismael Valenzuela (@aboutsecurity / @thinkredactblue)

import numpy as np
import matplotlib.pyplot as plt


categories = ['Damage Potential', 'Reproducibility', 'Exploitability', 'Affected Users','Discoverability']
categories = [*categories, categories[0]]


dropbox = [10, 5, 5, 2.5, 10]
kepass = [8, 5, 3, 6, 5]
evernote = [10, 7.5, 5, 2.5, 5]
printing = [5, 8, 5, 2.5, 8]
unifi = [10, 1, 2, 8, 0]

dropbox = [*dropbox, dropbox[0]]
kepass = [*kepass, kepass[0]]
evernote = [*evernote, evernote[0]]
printing = [*printing, printing[0]]
unifi = [*unifi, unifi[0]]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(dropbox))

plt.figure(figsize=(10, 10))
plt.subplot(polar=True)
plt.plot(label_loc, dropbox, label='Dropbox')
plt.plot(label_loc, kepass, label='KeePass')
plt.plot(label_loc, evernote, label='EverNote')
plt.plot(label_loc, printing, label='Printing Presses')
plt.plot(label_loc, unifi, label='Unifi UDM')
plt.title('DREAD comparison', size=20)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend()
plt.show()
```
